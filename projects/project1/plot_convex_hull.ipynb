# Carrega os arquivos dados

```
import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt("nuvem1.txt")
data2 = np.loadtxt("nuvem2.txt")
hull1 = np.loadtxt("fecho1.txt").astype(int)
hull2 = np.loadtxt("fecho2.txt").astype(int)
```

# Define a função para plotar a nuvem de pontos e o fecho convexo

```
def plot_hull(data_name, data, hull):
  fig, ax = plt.subplots()
  ax.set_title(data_name)
  ax.scatter(data[:, 0], data[:, 1])
  for i in range(len(hull)):
    if i != len(hull) - 1:
        plt.plot([data[hull[i]][0], data[hull[i + 1]][0]], [data[hull[i]][1], data[hull[i + 1]][1]], color="red")
    else:
        plt.plot([data[hull[i]][0], data[hull[0]][0]], [data[hull[i]][1], data[hull[0]][1]], color="red")
  plt.show()
```

# Fecho convexo dos arquivos "nuvem1.txt" e "nuvem2.txt"

```
plot_hull("Nuvem 1", data1, hull1)
```

```
plot_hull("Nuvem 2", data2, hull2)
```

# Fecho convexo das nuvens de pontos geradas aleatoriamente

```
point_size = [10**3, 5*10**3, 10**4, 5*10**4, 10**5, 5*10**5, 10**6]
for n in point_size:
  data = np.loadtxt("pointCloud{0}.txt".format(n))
  hull = np.loadtxt("convexHull{0}.txt".format(n)).astype(int)
  plot_hull("Nuvem {0}".format(n), data, hull)
```
