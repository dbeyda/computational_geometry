# Carrega os arquivos dados

```
import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt("nuvem1.txt")
data2 = np.loadtxt("nuvem2.txt")
tri1 = np.loadtxt("delaunay1.txt").astype(int)
tri2 = np.loadtxt("delaunay2.txt").astype(int)
```

# Define a função para plotar a nuvem de pontos e a triangulação

```
def plot_triangulation(data_name, data, tri):
  fig, ax = plt.subplots()
  ax.set_title(data_name)
  ax.scatter(data[:, 0], data[:, 1])
  for i in range(len(tri)):
    plt.plot([data[tri[i][0]][0], data[tri[i][1]][0]], [data[tri[i][0]][1], data[tri[i][1]][1]], color="red")
    plt.plot([data[tri[i][0]][0], data[tri[i][2]][0]], [data[tri[i][0]][1], data[tri[i][2]][1]], color="red")
    plt.plot([data[tri[i][1]][0], data[tri[i][2]][0]], [data[tri[i][1]][1], data[tri[i][2]][1]], color="red")
  plt.show()
```

# Triangulação dos arquivos "nuvem1.txt" e "nuvem2.txt"

```
print("Tamanho da nuvem 1: %d" % len(data1))
print("Quantidade de triângulos da triangulação: %d" % len(tri1))
plot_triangulation('Triangulação 1', data1, tri1)
```

```
print("Tamanho da nuvem 2: %d" % len(data2))
print("Quantidade de triângulos da triangulação: %d" % len(tri2))
plot_triangulation('Triangulação 2', data2, tri2)
```

# Teste do tempo de execução

```
point_size = [30, 70, 10**2, 3*10**2, 7*10**2, 10**3, 3*10**3, 7*10**3, 10**4]
sizes = [3, 7, 10, 30, 70, 100, 300, 700, 1000]
run_time = [0, 2, 4, 30, 109, 190, 1479, 8402, 16714]
             
plt.plot(point_size, [ms / 1000 for ms in run_time], "r")
plt.plot(point_size, [size**2/60000 for size in sizes], "b")
plt.xlabel('Número de pontos')
plt.ylabel('Tempo (s)')
plt.legend(["Delaunay","n**2"])
plt.grid(True)
plt.show()
```

# Triangulação das nuvens de pontos geradas aleatoriamente

```
point_size = [30, 70, 10**2, 3*10**2, 7*10**2, 10**3, 3*10**3, 7*10**3, 10**4]
for n in point_size:
  data = np.loadtxt("pointCloud{0}.txt".format(n))
  tri = np.loadtxt("delaunayTri{0}.txt".format(n)).astype(int)
  plot_triangulation("Nuvem {0}".format(n), data, tri)
```
